## Sensitivity of SF6 GWP

Here, we calculate the effect that dynamic characterization has on the SF6 end-of-life emissions from switchgears. Calculations according to [IPCC AR6, Chapter 7 and SM7](https://www.cambridge.org/core/product/identifier/9781009157896/type/book). Numerical values for Radiative Efficiencies (RE) originate from [Hodnebrog et al.](https://doi.org/10.1029/2019RG000691). Numerical values for the CO2 impulse response function originate from [Joos et al.](https://acp.copernicus.org/articles/13/2793/2013/acp-13-2793-2013.html).

```python
import numpy as np

M_air = 28.97
m_atmosphere = 5.135e18
RE_co2_kg = 1.33e-5 * M_air / 44.01 * 1e9 / m_atmosphere
RE_sf6_kg = 0.567 * M_air / 146.0607 * 1e9 / m_atmosphere

time_horizons = [70, 100]
co2_irf = lambda H: (
    0.2173 * H
    + 0.2240 * 394.4 * (1 - np.exp(-H / 394.4))
    + 0.2824 * 36.54 * (1 - np.exp(-H / 36.54))
    + 0.2763 * 4.304 * (1 - np.exp(-H / 4.304))
)

gwp_sf6 = {
    H: (RE_sf6_kg * 3200 * (1 - np.exp(-H / 3200))) / (RE_co2_kg * co2_irf(H))
    for H in time_horizons
}

relative_difference = (gwp_sf6[100] - gwp_sf6[70]) / gwp_sf6[100]

print(f"SF6 GWP70:  {gwp_sf6[70]:.0f}")
print(f"SF6 GWP100: {gwp_sf6[100]:.0f}")
print(f"Relative difference (100 vs 70): {relative_difference:.2%}")
```
