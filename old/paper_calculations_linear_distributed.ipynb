```python
def test_distributed_sums(components, distributed_data):
    original_totals = {}
    for year in components:
        if year != 2023:
            for component, value in components[year].items():
                original_totals[component] = original_totals.get(component, 0) + value

    distributed_sums = {component: 0 for component in original_totals.keys()}
    for year in distributed_data.values():
        for component, value in year.items():
            distributed_sums[component] += value

    all_passed = True
    for component, total in original_totals.items():
        if not round(distributed_sums[component], 1) == round(total, 1):
            print(f"Test failed for {component}: distributed sum {distributed_sums[component]} != original total {total}")
            all_passed = False

    if all_passed:
        print("All tests passed!")
    else:
        print("One or more tests failed.")
    return all_passed

# Usage:
test_result = test_distributed_sums(components, distributed_components)
```

```python
def aggregate_df(df, max_entries=None, cutoff=None):
    aggregated_df = pd.DataFrame()

    for scenario in df.index.get_level_values('Scenario').unique():
        scenario_df = df.xs(scenario, level='Scenario').copy()

        # Filter or limit the rows based on 'cutoff' or 'max_entries'
        if cutoff is not None:
            above_cutoff = scenario_df['GWI 100'] > cutoff
            other_sum = scenario_df[~above_cutoff]['GWI 100'].sum()
            scenario_df = scenario_df[above_cutoff]
        elif max_entries is not None and len(scenario_df) > max_entries:
            scenario_df = scenario_df.sort_values(by='GWI 100', ascending=False)
            other_sum = scenario_df.iloc[max_entries:]['GWI 100'].sum()
            scenario_df = scenario_df.iloc[:max_entries]
        else:
            other_sum = 0

        # Append the 'Other' row if there's something to aggregate
        if other_sum > 0:
            other_df = pd.DataFrame({'GWI 100': [other_sum]}, index=['other'])
            scenario_df = pd.concat([scenario_df, other_df])

        # Set the new MultiIndex without resetting and reindexing
        scenario_df.index = pd.MultiIndex.from_product([[scenario], scenario_df.index], names=['Scenario', 'Activity Name'])

        # Append to the aggregated DataFrame
        aggregated_df = pd.concat([aggregated_df, scenario_df])

    return aggregated_df
```

```python
# Initialize a dictionary to hold the summed values
material_sums = {}

# Iterate through the data to sum the materials across all components
for scenario, components in results.items():
    for component, materials in components.items():
        for material, value in materials.items():
            if material not in material_sums:
                material_sums[material] = {}
            if scenario not in material_sums[material]:
                material_sums[material][scenario] = 0
            material_sums[material][scenario] += value

# Convert the summed materials dictionary to a DataFrame
df_material_sums = pd.DataFrame.from_dict(material_sums, orient='index')
```
