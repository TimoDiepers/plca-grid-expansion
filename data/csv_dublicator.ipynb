```python
import pandas as pd
import os

# Verzeichnis mit den CSV-Dateien
directory = '.'

# Alle CSV-Dateien im Verzeichnis durchgehen
for filename in os.listdir(directory):
    if filename.endswith('.csv') and not filename.endswith('_future.csv') and not filename == 'sankey_data.csv':
        filepath = os.path.join(directory, filename)
        
        # CSV-Datei einlesen (mit ; als Trennzeichen)
        df = pd.read_csv(filepath, sep=';')
        
        # 'amount'-Spalte halbieren (dummy Wert)
        df['amount'] = df['amount'] * 0.5 # for now, just a dummy value
        
        # Neue Datei schreiben
        new_filename = filename.replace('.csv', '_future.csv')
        new_filepath = os.path.join(directory, new_filename)
        df.to_csv(new_filepath, sep=';', index=False)

        print(f"Erstellt: {new_filename}")
```

```python
# Alle _future.csv-Dateien durchgehen und löschen
for filename in os.listdir(directory):
    if filename.endswith('_future.csv'):
        filepath = os.path.join(directory, filename)
        os.remove(filepath)
        print(f"🗑️ Gelöscht: {filename}")
```
